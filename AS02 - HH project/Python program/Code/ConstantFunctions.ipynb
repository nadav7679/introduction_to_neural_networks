```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
def nVars(V):
    
    alpha = 0.01*(V+50)/(1-np.exp(-(50+V)/10))
    beta = 0.125*np.exp(-(V+60)/80)
    x = alpha/(alpha + beta);
    y = 1/(alpha + beta);
    
    return x,y

def mVars(V):
    
    alpha = 0.1*(35+V)/(1-np.exp(-(35+V)/10))
    beta = 4*np.exp(-0.0556*(60+V))
    x = alpha/(alpha + beta);
    y = 1/(alpha + beta);
    
    return x,y

def hVars(V):
    
    alpha = 0.07*np.exp(-0.05*(V+60))
    beta = 1/(1+np.exp(-0.1*(V+30)))
    x = alpha/(alpha + beta);
    y = 1/(alpha + beta);
    return x,y
```

```python
domain = np.arange(-100,20,0.1)
n = np.zeros(len(domain),dtype=float)
m = np.zeros(len(domain),dtype=float)
h = np.zeros(len(domain),dtype=float)


for i in range(len(domain)):
    n[i]= nVars(domain[i])[1]
    m[i]= mVars(domain[i])[1]
    h[i]= hVars(domain[i])[1]


plt.plot(domain,n, label = "n")
plt.plot(domain,m,label="m")
plt.plot(domain,h,label="h")
plt.legend()
plt.xlabel("Voltage [mV]")
plt.ylabel("Time constant")

plt.grid()
plt.savefig("Time constant Graph.png")
```
